```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.random.sample(100)
y = 2*x + 1 + np.random.sample(len(x))

z = np.polyfit(x,y, 1)
p = np.poly1d(z)

plt.scatter(x,y)
plt.plot(x,p(x))

plt.show()
```

```python
data = np.identity(5)
sns.heatmap(data)

plt.show()
```

```python
z , w = np.random.sample(100), np.random.sample(100)

data = pd.DataFrame(np.array([x,y,z,w]).T)
data.head()
sns.heatmap(data.corr())

plt.show()
```
