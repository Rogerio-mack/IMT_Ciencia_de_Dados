# **PCA e SVD (Singular Value Decomposition)**

Como vimos o PCA é uma técnica de análise de dados que visa encontrar uma projeção dos dados em seus componentes principais, de modo que esses componentes(ortogonais) capturem a maior variabilidade nos dados originais. **SVD (Singular Value Decomposition)**, por outro lado, é uma técnica algébrica mais geral que descompõe uma matriz retangular em três outras matrizes. Ela não está especificamente voltada para análise de dados, mas é amplamente usada em diversas aplicações, como álgebra linear numérica, processamento e compactação de imagens, e no aprendizado de máquina como ferramenta de redução de dimensionalidade, sendo o exemplo mais famoso o uso do SVD para vencer o **NetFlix Prize** (2006-2009) em que os dados envolviam 100 milhões (100.480.507 para ser exato, mas o que é apenas cerca 1% do total de casos possíveis) de itens de classificação, para 480.189 usuários e 17.770 filmes. De qualquer forma, para os nossos propósitos, basta entender que o SVD funciona do mesmo modo que o PCA como um redutor de dimensionalidade. Na verdade o PCA é um caso particular do SVD em que os componentes são ortogonais - o SVD não tem essa restrição.

```python
selected_names = ['Naomi Watts','Winona Ryder','Tiger Woods','Vladimir Putin','Serena Williams','Angelina Jolie']
selected_names.sort()
index_selection = [i for i in range(len(faces.target)) if faces.target_names[faces.target[i]] in selected_names]

faces.data = faces.data[ index_selection ]
faces.target = faces.target[ index_selection ]
faces.images = faces.images[ index_selection ]
faces.target_names = faces.target_names[ faces.target ]
```

```python
fig, ax = plt.subplots(4, 6, figsize=(12, 10))
axi = ax.flatten()

for i in range(len(axi)):
  axi[i].imshow(faces.images[i], cmap='gray')
  axi[i].set(xticks=[], yticks=[], xlabel=faces.target_names[i])
```
