e você pode repetir isso para cada um dos setores de modo iterativo: 

````
g = company.groupby('Setor')

my_dfs = { }

for key in g.groups.keys():
  temp = g.get_group(key).drop(columns="Setor").reset_index(drop=True)
  my_dfs[key] = temp

for key, data in my_dfs.items():
  print('DataFrame: ', key)
  display( data.head() )
  data.to_csv(key + '.csv',index=None)
  data.to_json(key + '.json')
  print( f'memory usage: {data.memory_usage().sum() / 1024:.1f}KB' )

````

Ou ainda empregar o comando `exec`.

````
g = company.groupby('Setor')

for key, group in g:
  print('DataFrame: ', key)
  exec( key + ' = group.drop(columns="Setor").reset_index(drop=True)' )
  exec( 'display(' + key + '.head())')
  exec( 'print( "memory usage: " , np.round(' + key + '.memory_usage().sum() / 1024 , 1) , "KB") ' )
  exec( key + '.to_csv("' + key + '.csv", index=None)')
  exec( key + '.to_json("' + key + '.json")')
  print()
````

## Q3. `agg()`, `mode()`

Veja os exemplos abaixo de como empregar a função `agg()` e como criar uma função nova para uso dentro do `agg()`. 

````
df
````

````
def my_mode(x): # x.mode() é um método que tornamos aqui como uma função para poder ser empregada com agg()
  return x.mode()
````

````
my_mode(df['order'])
````

````
df.groupby('class')['order'].agg(my_mode)
````

````
import numpy as np

df.groupby('class')[['order','max_speed']].agg({'order': my_mode, 'max_speed': np.mean})
````

Empregue os exemplos para fornecer os valores de moda (atributos categóricos) e médios (atributos numéricos) para todos os valores de `company` por Setor.

````
import numpy as np

company.groupby('Setor')[['Sexo', 'Grau_Instrucao', 'Idade', 'Anos_Experiencia', 
                    'Estado', 'Salario']].agg({'Sexo': (lambda x: x.mode()[0]), # alternativa a my_mode()
                                              'Grau_Instrucao': my_mode,
                                              'Idade': np.mean,
                                              'Anos_Experiencia': np.mean,
                                              'Estado': my_mode,
                                              'Salario': np.mean}).reset_index()
````

## Q4. `map()`

Empregue `map()` para criar um atributo `Codigo_instrucao` atribuindo valores inteiros $1$ a $n$ do menor para o maior `Grau_Instrucao` (ordem alfabética).



````
grau_instrucao = sorted( list( company.Grau_Instrucao.unique() ) )
grau_instrucao
````

````
instrucao_map = {}
for a, b in zip(grau_instrucao,np.arange(1,len(grau_instrucao)+1)): # black, blue, green, yellow, magenta
  instrucao_map[a] = b

instrucao_map
````

````
company['Codigo_Instrucao'] = company.Grau_Instrucao.map(instrucao_map)
company.head()
````
