Neste laboratório:

1. Reveja os modelos de regressão linear com statsmodel
2. Entenda os efeitos da normalização


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('http://meusite.mackenzie.br/rogerio/data_load/regressao_preco_imoveis.xlsx')
df.head()
```

```python
df = df.drop(columns='bairro')
df.head()
```

# Q1.

Considere somente os dados numéricos, sem transformações. Qual o R2 obtido da regressão linear para estimativa do preço dos imóveis?

```python
import statsmodels.formula.api as sm

lm = sm.ols(formula='preco ~ areaM2 + suites + dormitorios + banheiros + vagas', data=df)
lm = lm.fit()
print(lm.summary())
```

# Q2.

# Q3.

Considere somente os dados numéricos, sem transformações. Qual o R2 obtido somente com os coeficientes significativos?

```python
import statsmodels.formula.api as sm

lm = sm.ols(formula='preco ~ areaM2 + suites + dormitorios + vagas - 1', data=df)
lm = lm.fit()
print(lm.summary())
```

# Q4.

Considere somente os dados numéricos, sem transformações. Considerando somente os coeficientes significativos da regressão linear, qual o preço estimado de um imóvel de 134 m2, uma suite, 4 dormitórios e 3 vagas de garagem?


```python
df_case = pd.DataFrame()
df_case['areaM2'] = [134]
df_case['suites'] = [1]
df_case['dormitorios'] = [4]
df_case['vagas'] = [3]

df_case
```

```python
preco = lm.predict(df_case)
print()
print(f'Preco estimado: {preco[0]:.2f}')
```

# Qb.

Normalize os dados, exceto o preço, agora pela normalização **MinMAx** (faça sem empregar o scikit-learn) e refaça a regressão linear com os preditores significativos. Qual a estimativa do preço agora?


```python
df_scaled = ( df.drop(columns='preco') - df.drop(columns='preco').min() ) / ( df.drop(columns='preco').max() - df.drop(columns='preco').min() )
df_scaled = pd.concat([df_scaled,df[['preco']]],axis=1)

df_scaled.head()


```

```python
import statsmodels.formula.api as sm

lm = sm.ols(formula='preco ~ areaM2 + suites + dormitorios + vagas - 1', data=df_scaled)
lm = lm.fit()
print(lm.summary())
```

```python
df_case_scaled  = (df_case - df.drop(columns='preco').min()) / ( df.drop(columns='preco').max() - df.drop(columns='preco').min())

display(df_case_scaled)

preco = lm.predict(df_case_scaled)
print()
print(f'Preco estimado: {preco[0]:.2f}')
```
