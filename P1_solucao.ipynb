```
Q9A. (Fácil) Leia agora a base de dados 'df_train.csv'. LEIA NOVAMENTE, mesmo que você já tenha empregado essa base antes carregue-a
novamente. Você vai agora construir as variáveis preditoras para um modelo de classificação de Type_of_glass. O preço não será empregado como variável preditora, assim como outras variáveis que não FAZEM SENTIDO COMO PREDITORAS. Crie então conjunto de dados X a partir das seguintes transformações nos dados originais seguinte esta ordem\:
\n a. EXCLUA os valores ausentes.
\n b. EXCLUA os valores de 'price'.
\n c. Reclassifique os tipos de vidro para reciclagem em 3 categorias\: C \= Vidros de Construção; V \= Vidros de Veículos; O \= Outros (DICA\: você pode empregar a função  'replace()' do Pandas e note que vidros de faróis são de veículos).
\n d. Faça o HOT ENCODE dos atributos categóricos, considerando o valor mais frequente para ser excluído evitando-se a multicolinearidade e atribuindo valores TRUE ou FALSE.
\n e. EXCLUA os demais atributos não preditores.
\n f. NORMALIZE (minmax) os dados obtidos, incluindo os valores booleanos criados no hot encode acima (eles são normalizados do mesmo modo).
\n\nQual a soma de todos valores obtidos?
{
= 916.81
~ 1916.24
~ 8098.22
~ 178899.32
~ 8898.45
}

```

````python
df = pd.read_csv(path + 'df_train.csv')

df = df.dropna()

df.drop(columns='price',inplace=True)

# seu código
df['Type_of_glass'] = df['Type_of_glass'].replace(1,'C')
df['Type_of_glass'] = df['Type_of_glass'].replace(2,'C')
df['Type_of_glass'] = df['Type_of_glass'].replace(3,'V')
df['Type_of_glass'] = df['Type_of_glass'].replace(4,'V')
df['Type_of_glass'] = df['Type_of_glass'].replace(5,'O')
df['Type_of_glass'] = df['Type_of_glass'].replace(6,'O')
df['Type_of_glass'] = df['Type_of_glass'].replace(7,'V')

df['D'] = df.Mg == 'D'
df['B'] = df.Mg == 'B'
df['A'] = df.Mg == 'A'

X = df.drop(columns=['Id number', 'receipt name', 'receipt', 'Type_of_glass', 'Mg'])
print(X.head())
X = X / X.max()
print(X)
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# Entradas e Saídas
X = X
y = df['Type_of_glass']

# Separando conjuntos de Treinamento e Teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=1)

# Definição do modelo
clf1 = LogisticRegression(max_iter=10000)
clf2 = KNeighborsClassifier(5)
clf3 = KNeighborsClassifier(3)

# Treinamento
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
clf3.fit(X_train,y_train)

# Acuracidade
print(f'Logistic, {clf1.score(X_test, y_test):.4f}')
print(f'Knn 5, {clf2.score(X_test, y_test):.4f}')
print(f'Knn 3, {clf3.score(X_test, y_test):.4f}')
````
